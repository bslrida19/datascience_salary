```python
df=df[df['Salary Estimate'] != '-1']
```

```python
df['Salary Estimate'].value_counts()
```

```python
def cleanSALARY(x):
    x=x.split(' ')
    z=x[0].split('-')
    z[0]=z[0].replace('$','')
    z[0]=z[0].replace('K','')
    z[1]=z[1].replace('$','')
    z[1]=z[1].replace('K','')
    y=(int(z[0])+int(z[1]))/2
    
    return y
    
def minSALARY(x):
    x=x.split(' ')
    z=x[0].split('-')
    z[0]=z[0].replace('$','')
    z[0]=z[0].replace('K','')

    return z[0]
    
def maxSALARY(x):
    x=x.split(' ')
    z=x[0].split('-')
    z[1]=z[1].replace('$','')
    z[1]=z[1].replace('K','')
    
    return z[1]
    
    
```

```python
df['Average slary']=df['Salary Estimate'].apply(cleanSALARY)
```

```python
df['Min slary']=df['Salary Estimate'].apply(minSALARY)
```

```python
df['Max slary']=df['Salary Estimate'].apply(maxSALARY)
```
